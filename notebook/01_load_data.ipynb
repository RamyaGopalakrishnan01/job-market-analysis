```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Count jobs by location
top_locations = df['location'].value_counts().head(10)
print(top_locations)

# Plot it
plt.figure(figsize=(10,6))
sns.barplot(x=top_locations.values, y=top_locations.index, palette="viridis")
plt.title("Top 10 Locations Hiring Data Analysts")
plt.xlabel("Number of Jobs")
plt.ylabel("Location")
plt.show()
```

```python
# Count jobs by industry
top_industries = df['industry'].value_counts().head(10)

# Plot it
plt.figure(figsize=(10,6))
sns.barplot(x=top_industries.values, y=top_industries.index, palette="coolwarm")
plt.title("Top 10 Industries Hiring Data Analysts")
plt.xlabel("Number of Jobs")
plt.ylabel("Industry")
plt.show()
```

```python
# Histogram of company ratings
plt.figure(figsize=(8,5))
sns.histplot(df['rating'], bins=10, kde=True, color='green')
plt.title("Distribution of Company Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
# Remove $ and K, and text in parentheses
salary_clean = df['salary_estimate'].str.replace(r'[\$,K]', '', regex=True)
salary_clean = salary_clean.str.split(' ').str[0]  # take only the first part before any text like (Glassdoor est.)

# Now split by '-' safely
df['min_salary'] = salary_clean.str.split('-').str[0].astype(float)

# For max salary, if no '-', use min_salary
df['max_salary'] = salary_clean.str.split('-').str[1]  # might be NaN
df['max_salary'] = df['max_salary'].fillna(df['min_salary']).astype(float)

# Average salary
df['avg_salary'] = (df['min_salary'] + df['max_salary']) / 2

# Check first few rows
df[['salary_estimate', 'min_salary', 'max_salary', 'avg_salary']].head(10)
```

```python
# Top 10 locations by number of jobs
top_locations = df['location'].value_counts().head(10).index

# Calculate average salary per location
location_salary = df[df['location'].isin(top_locations)].groupby('location')['avg_salary'].mean().sort_values(ascending=False)

# Plot
plt.figure(figsize=(10,6))
sns.barplot(x=location_salary.values, y=location_salary.index, palette="magma")
plt.title("Average Salary by Top 10 Locations")
plt.xlabel("Average Salary (K)")
plt.ylabel("Location")
plt.show()
```

```python
# Top 10 industries by number of jobs
top_industries = df['industry'].value_counts().head(10).index

# Calculate average salary per industry
industry_salary = df[df['industry'].isin(top_industries)].groupby('industry')['avg_salary'].mean().sort_values(ascending=False)

# Plot
plt.figure(figsize=(10,6))
sns.barplot(x=industry_salary.values, y=industry_salary.index, palette="cool")
plt.title("Average Salary by Top 10 Industries")
plt.xlabel("Average Salary (K)")
plt.ylabel("Industry")
plt.show()
```

```python
# Scatter plot: rating vs avg_salary
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='rating', y='avg_salary', alpha=0.6)
plt.title("Company Rating vs Average Salary")
plt.xlabel("Company Rating")
plt.ylabel("Average Salary (K)")
plt.show()
```

```python
# Top 10 companies by number of job postings
top_companies = df['company_name'].value_counts().head(10)

# Plot
plt.figure(figsize=(10,6))
sns.barplot(x=top_companies.values, y=top_companies.index, palette="plasma")
plt.title("Top 10 Companies Hiring Data Analysts")
plt.xlabel("Number of Job Postings")
plt.ylabel("Company Name")
plt.show()
```

```python
# Take only top 10 companies
top_companies_list = top_companies.index

# Calculate average salary
company_salary = df[df['company_name'].isin(top_companies_list)].groupby('company_name')['avg_salary'].mean().sort_values(ascending=False)

# Plot
plt.figure(figsize=(10,6))
sns.barplot(x=company_salary.values, y=company_salary.index, palette="coolwarm")
plt.title("Average Salary for Top 10 Companies Hiring")
plt.xlabel("Average Salary (K)")
plt.ylabel("Company Name")
plt.show()
```

```python
# Create summary table
company_summary = df[df['company_name'].isin(top_companies_list)].groupby('company_name').agg(
    num_jobs=('job_title','count'),
    avg_salary=('avg_salary','mean')
).sort_values(by='num_jobs', ascending=False)

company_summary
```
